```python
import requests
```

```python

urls = [
    "http://www.philippe-fournier-viger.com/spmf/datasets/mushroom_utility_SPMF.txt",
    "https://www.philippe-fournier-viger.com/spmf/datasets/chess_utility_spmf.txt",
    "https://www.philippe-fournier-viger.com/spmf/datasets/connect_utility_spmf.txt",
    "https://www.philippe-fournier-viger.com/spmf/datasets/accidents_utility_spmf.txt",
    "https://www.philippe-fournier-viger.com/spmf/datasets/kosarak_utility_spmf.txt",
    "https://www.philippe-fournier-viger.com/spmf/datasets/chainstore.txt",    
]
filenames = [
    "mushroom_utility",
    "chess_utility",
    "connect_utility",
    "accidents_utility",
    "kosarak_utility",
    "chainstore_utility"   
]

path = "datasets/"
for i in range(6):
    r = requests.get(urls[i], allow_redirects=True)
    open(path + filenames[i] + ".txt", 'wb').write(r.content)
```

```python
import tracemalloc
 
```

```python
# code or function for which memory
# has to be monitored
def app(n):
    lt = []
    for i in range(0, n):
        lt.append(i)
 
# starting the monitoring
tracemalloc.start()
 
# function call
app()
 
# displaying the memory
usage = tracemalloc.get_traced_memory()
total = usage[1] - usage[0]
kb = total / 1024
mb = kb / 1024
print(f'memory used in KiB: {kb:.2f}')
print(f'memory used in MiB: {mb:.2f}')
 
# stopping the library
tracemalloc.stop()
```

```python
import matplotlib.pyplot as plt
```

```python
x_chess = [600, 500, 400, 300, 200]
cor_chess = [0.86,0.78,0.7]
```

```python

for i in range(3):
    fig, ax = plt.subplots()
    ax.plot(x_chess[::-1],chess_record_time[i], marker='o')
    ax.set_xticks(x_chess)
    ax.set_xticklabels(x_chess[::-1])
    ax.set_ylim(0, max(chess_record_time[i])*1.5)
    ax.set_title(f"Chess (minCorr = {cor_chess[i]})")
    ax.set_xlabel("Minimum utility (in 1000s)")
    ax.set_ylabel("Runtimes(s)")
    ax.grid(True)
```
